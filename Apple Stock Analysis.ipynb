```python
data['%-change'] = data['Close'].pct_change()                  # using inbuilt methods
data.head()
```

```python
data.head()
```

```python
import numpy as np
data['log_returns'] = np.log(data['Close']/data['Close'].shift())
```

```python
data.head()
```

```python
data['Close'].shift(2)              # it shifts
```

```python
data['log_returns'].std()
```

```python
#sigma(annually) = sigma(daily) * under_root(252)  (252 trading days per year)
volatitily = data['log_returns'].std()*252**0.5
volatitily
```

```python
str_vol = str(round(volatitily,4)*100)
str_vol
```

```python
fig, ax = plt.subplots()
data["log_returns"].hist(ax=ax, bins = 50, alpha = 0.6, color = 'blue')
ax.set_xlabel("Log return")
ax.set_ylabel("Freq of Log return")
ax.set_title("AAPL volatility : " + str_vol + "%")
```
